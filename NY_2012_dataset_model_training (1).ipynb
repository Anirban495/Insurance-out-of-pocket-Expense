```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("cleaned_2012_dataset.csv", index_col=0)
```

```python
df
```

```python
df4=df.copy()
```

```python
# Performing frequency encoding, althogh I am not conviced that this should be approach because of huge difference in counts 

County_frequency_df = df4.groupby('Hospital County').size().reset_index(name='County_Frequency')

FN_frequency_df = df4.groupby('Facility Name').size().reset_index(name='Facility_Name_Frequency')

CCS_frequency_df = df4.groupby('CCS Procedure Description').size().reset_index(name='CCS_Procedure_Frequency')

df4 = df4.merge(County_frequency_df, on='Hospital County', how='left')

df4 = df4.merge(FN_frequency_df, on='Facility Name', how='left')

df4 = df4.merge(CCS_frequency_df, on='CCS Procedure Description', how='left')
```

```python
df4
```

```python
# Performing One hot encoding 
one_hot_encoded = pd.get_dummies(df4[['Age Group','Ethnicity', 'Race','Gender','APR DRG Description']],drop_first=True, dtype=float)

df4 = pd.concat([df4, one_hot_encoded], axis=1)
```

```python
# dropping encoded columns
df4.drop(columns=['Facility Name','Age Group','Ethnicity', 'Race','Gender','APR DRG Description'], axis=1, inplace=True)
```

```python
# Performing Ordinal Encoding

df4['Type of Admission']=df4['Type of Admission'].map({'Emergency':4, 'Elective':1, 'Urgent':2, 'Not Available':0, 'Newborn':3})

df4['APR Risk of Mortality']=df4['APR Risk of Mortality'].map({'Moderate':2, 'Minor':1, 'Major':3, 'Extreme':4})

df4['APR Medical Surgical Description']=df4['APR Medical Surgical Description'].map({'Medical':1, 'Surgical':2})
```

```python
df4['Abortion Edit Indicator'] = df4['Abortion Edit Indicator'].map({'N':0, 'Y':1})

df4['Emergency Department Indicator'] = df4['Emergency Department Indicator'].map({'N':0, 'Y':1})
```

```python
df4
```

```python
df4['Zip Code - 3 digits'].unique()
```

```python
df4['Zip Code - 3 digits'].replace({'OOS':'99'}, inplace=True)
```

```python
df4['Zip Code - 3 digits'] = df4['Zip Code - 3 digits'].astype(int)
```

```python
X = df4.drop(columns=['Hospital County','CCS Procedure Description','Payment Typology 1','Facility_Name_Frequency', 'Total Charges'], axis=1)
y = df4['Total Charges']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.20, random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor

RF_reg =  RandomForestRegressor(n_estimators=100)

RF_reg.fit(X_train, y_train)

y_pred = RF_reg.predict(X_test)

print("Test set results")

print('mse: ', mean_squared_error(y_test, y_pred))

print('r2: ',r2_score(y_test, y_pred))

print('mae: ',mean_absolute_error(y_test, y_pred))

print('rmse: ',mean_squared_error(y_test, y_pred,squared=False))

print("--------------------------------------------------------")

print('Train set results')
y_train_pred = RF_reg.predict(X_train)
print('mse: ', mean_squared_error(y_train, y_train_pred))
print('r2: ',r2_score(y_train, y_train_pred))
print('mae: ',mean_absolute_error(y_train,y_train_pred))
print('rmse: ',mean_squared_error(y_train, y_train_pred,squared=False))
```

```python
from xgboost import XGBRegressor

XGB_reg = XGBRegressor()

XGB_reg.fit(X_train, y_train)

y_pred = XGB_reg.predict(X_test)

print("Test set results")

print('mse: ', mean_squared_error(y_test, y_pred))

print('r2: ',r2_score(y_test, y_pred))

print('mae: ',mean_absolute_error(y_test, y_pred))

print('rmse: ',mean_squared_error(y_train, y_train_pred,squared=False))

print("--------------------------------------------------------")
print('Train set results')
y_train_pred = XGB_reg.predict(X_train)
print('mse: ', mean_squared_error(y_train, y_train_pred))
print('r2: ',r2_score(y_train, y_train_pred))
print('mae: ',mean_absolute_error(y_train,y_train_pred))
print('rmse: ',mean_squared_error(y_test, y_pred,squared=False))
```

```python
County_counts = df5['Hospital County'].value_counts().sort_values()

County_list = County_counts.index.to_list()

ordinal_mapping = {county: i + 1 for i, county in enumerate(County_list)}

df5['County_Ordinal'] = df5['Hospital County'].map(ordinal_mapping)
```

```python
CCS_pro_counts = df5['CCS Procedure Description'].value_counts().sort_values()

CCS_pro_list = CCS_pro_counts.index.to_list()

ordinal_mapping = {CCS_pro: i + 1 for i, CCS_pro in enumerate(CCS_pro_list)}

df5['CCS_pro_Ordinal'] = df5['CCS Procedure Description'].map(ordinal_mapping)
```
